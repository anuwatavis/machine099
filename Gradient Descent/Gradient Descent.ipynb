```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```


# Example 1 - A Simple cost fucntion
## $f(x) = x^2 + x +1$

```python
def f(x):
    return x**2 + x + 1
```

```python
#Make Data
x_1 = np.linspace(start=-3, stop=3, num=500)
x_1
```

```python
#Plot
plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize =16)
plt.plot(x_1, f(x_1))
plt.show()
```

# Slope & Derivatives
challenge : Create a python function for the derivative of $f(x)$ called df(x)

```python
def df(x):
    return 2*x + 1
```

```python
#Plot function and derivative side by side
plt.figure(figsize=[5, 15])
# 1 Chart : Cost Function
plt.subplot(2, 1, 1) # row column index

plt.xlim(-3, 3)
plt.ylim(0, 8)
plt.xlabel('X', fontsize = 16)
plt.ylabel('f(x)', fontsize =16)
plt.title('Cost Function', fontsize=17)
plt.plot(x_1, f(x_1), color='blue', linewidth=3)

# Chart: Derivative
plt.subplot(2, 1, 2)
plt.xlabel('X', fontsize = 16)
plt.ylabel('df(x)', fontsize =16)
plt.title('Slope of the cost function', fontsize=17)
plt.xlim(-2, 3)
plt.ylim(-3, 6)
plt.plot(x_1, df(x_1), color='skyblue', linewidth=3)
plt.show()
```

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1

for n in range(30):
    previous_x = new_x
    gradient  = df(previous_x)
    new_x = previous_x - step_multiplier * gradient 
    
print('Local minimum occirs at : ', new_x)
print('Slop of df(x) value at this point is', df(new_x))
print('f(x) value or cost at this point is :', f(new_x))
```
